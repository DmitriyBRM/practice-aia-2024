````python
import random
from random import choice


food = ['окрошку', 'мясо(говядина)', 'корм', 'колбасу', 'сосиски', 'листья салата', 'окуня', 'щуку', 'икру семги', 'хамон', 'фуагра', 'ботаргу']
move = None
hungry = 3


def tamePet(x):
    petType = input('введите тип животного: ')
    petName = input('введите имя животного: ')
    petAge = input('введите возраст живтоного: ')

    print(f'отлично! ваш питомец {petName} ({petType}) возраста {petAge} приручен.')

    pet = {'name': petName, 'type': petType, 'age': 'petAge'}

    return pet


def feedPet(pet):
    print('хотите покормить питомца чем-то конркетным?')

    action = int(input('1 - да |2 - нет'))

    if action == 1:
        fFeed = input('введите, чем вы хотите покормить питомца')

    elif action == 2:
        fFeed = choice(food)

    print(f'отлично! ваш питомец покушал {fFeed}')


def rock_paper_scissors():

    def f(x, y, a, b):
        if x == 'камень' and y == 'ножницы':
            a+= 1

        if x == 'камень' and y == 'бумага':
            b+= 1

        if x == 'ножницы' and y == 'камень':
            b+= 1

        if x == 'ножницы' and y == 'бумага':
            a+= 1

        if x == 'бумага' and y == 'ножницы':
            b+= 1

        if x == 'бумага' and y == 'камень':
            a+= 1

        if x == 'камень' and y == 'камень':
            a+= 1
            b+= 1

        if x == 'ножницы' and y == 'ножницы':
            a+= 1
            b+= 1

        return a, b


    hod = ['камень', 'ножницы','бумага']
    a, b = 0,0

    while a < 3 and b < 3:

        hod_chelovek = input('введите ваш ход:')

        print('_____________________________________')

        ran = random.choice(hod)
        r = (f(ran, hod_chelovek, a, b))

        print('питомец:', ran)
        print('игрок:', hod_chelovek)
        print('счет:', r)

        a, b = r

    else:
        if a == 3:
            print('победил питомец')

        elif b == 3:
            print('вы победили')


while move != 6:

    print('списсок команд')
    print('1) приручить питомца')
    print('2) покормить питомца')
    print('3) поиграть с питомцем')
    print('5) уложить спать')
    print('6) выход')

    move = int(input('введите команду: '))

    if move == 1:
        tamePet(None)
        hungry = 3

    elif move == 6:
        print('до скорых встреч!')

    elif move == 2:
        feedPet(None)
        hungry -= 1
        print(f'голод питомца: {hungry}')

    elif move == 3:
        if hungry >= 3:
            print('ваш питомец голоден. сначала покормите его!')

        else:
            hungry += 1

            print('хотите поиграть с питомцем в камень/ножницы/бумагу ?')
            print('1 - да |2 - нет')

            action = int(input())

            if action == 1:
                rock_paper_scissors()

            else:
                print('окс')
````

Для данной задачи используйте: https://pokeapi.co/


````python
import requests
import random

pokemons_comand = {}


def send_request(url):

    response = requests.get(url)

    if response.status_code == 200:
        return response.json()


def pokemon_info(url):
    data = send_request(url)
    pokemon_type = data['types'][0]['type']['name']
    height = data['height']
    weight = data['weight']
    abilities = []
    amount_abilities = len(data['abilities'])

    for number_ability in range(amount_abilities):
        abilities.append(data['abilities'][number_ability]['ability']['name'])

    information = {'type': pokemon_type, 'height': height, 'weight': weight, 'abilities': abilities}

    return information

def set_pokemon(url, name):
    if name not in pokemons_comand.keys():
        pokemons_comand[name] = pokemon_info(url)
    else:
        print('такой покемон уже есть в вашей команде!')

def fight(name1, name2):

    health1 = random.randrange(10,20)
    health2 = random.randrange(10,20)

    while health1 > 0 and health2 > 0:

        print('__________________________________________')
        print(f'у покемона {name1} здоровье {health1}')
        print(f'у покемона {name2} здоровье {health2}')
        print('__________________________________________')

        damage1 = random.randrange(5)
        damage2 = random.randrange(5)

        print(f'покемон {name1} наносит {damage1}')
        health2 -= damage1

        print(f'покемон {name2} наносит {damage2}')
        health1 -= damage2

    if health1 <= 0:
        print(f'{name1} проиграл')
    elif health2 <= 0:
        print(f'{name2} проиграл')




move = None

while move != 6:
    print('1) добавить покемона в команду')
    print('2) удалить покемона из команды ')
    print('3) посмотреть информацию о покемонах в команде')
    print('4) найти покемона')
    print('5) устроить тренировочный бой')
    print('6) выход')
    move  = int(input('введите команду: '))
    if move  == 1:
        name = input('введите имя покемона: ')
        url = f'https://pokeapi.co/api/v2/pokemon/{name}/'
        set_pokemon(url, name)

    elif move == 2:
        neme = input('введите имя покемона для удаления: ')
        del pokemons_comand[name]

    elif move == 3:
        print(pokemons_comand)

    elif move == 4:
        name = input('введите имя покемона: ')
        print(pokemons_comand[name])

    elif move == 5:

        print(pokemons_comand)

        name1 = input(f'введите имя 1 покемона')
        if name1 in pokemons_comand:
            name2 = input(f'введите имя 2 покемона')
            if name2 in pokemons_comand:
                fight(name1, name2)
            else:
                print('в вашей команде нет такого покемона')
        else:
            print('в вашей команде нет такого покемона')


    elif move == 6:
        break









````
